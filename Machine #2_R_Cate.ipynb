```python
from sklearn import tree
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join("PUF_FinalCSV.csv"))
df = df.fillna(0)
df.head()
```

```python
columns = []
for col in df.columns: 
    if col not in ['Q37_WEEKS_WORKED','PUF_ID','final_weight']:
        columns.append(col)
```

```python
df['Q3_Q3_A_R']=df.Q3_Q3_A_R.replace(to_replace={1.0:0})
df['Q3_Q3_A_R']=df.Q3_Q3_A_R.replace(to_replace={2.0:1})
df['Q3_Q3_B_R']=df.Q3_Q3_A_R.replace(to_replace={1.0:0})
df['Q3_Q3_B_R']=df.Q3_Q3_A_R.replace(to_replace={2.0:1})
df['Q3_Q3_C_R']=df.Q3_Q3_A_R.replace(to_replace={1.0:0})
df['Q3_Q3_C_R']=df.Q3_Q3_A_R.replace(to_replace={2.0:1})
df['Q3_Q3_D_R']=df.Q3_Q3_A_R.replace(to_replace={1.0:0})
df['Q3_Q3_D_R']=df.Q3_Q3_A_R.replace(to_replace={2.0:1})
df['Q3_Q3_E_R']=df.Q3_Q3_A_R.replace(to_replace={1.0:0})
df['Q3_Q3_E_R']=df.Q3_Q3_A_R.replace(to_replace={2.0:1})

#Q2_COL_CREDIT
df['Q2_COL_CREDIT']=df.Q2_COL_CREDIT.replace(to_replace={1.0:0})
df['Q2_COL_CREDIT']=df.Q2_COL_CREDIT.replace(to_replace={-3.0:0})

#Q5_COL_ENROLL
df['Q5_COL_ENROLL']=df.Q5_COL_ENROLL.replace(to_replace={1.0:0})

#Q9_CERT_LIC
df['Q9_CERT_LIC']=df.Q9_CERT_LIC.replace(to_replace={1.0:0})
df['Q9_CERT_LIC']=df.Q9_CERT_LIC.replace(to_replace={2.0:1})
```

```python
df.head(25)
```

```python
data = df[columns]
data.head()
```

```python
#data.columns = data.columns.astype('category')
for col in columns: 
    data[col]=pd.Categorical(data[col])
    
```

```python
data.head()
```

```python
data.dtypes
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
feature_names = data.columns
```

```python
target = df["Q37_WEEKS_WORKED"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
